## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate(array, direction):
    if direction == 'left':
        return [list(row) for row in zip(*[row[::-1] for row in array])]
    else:
        return [list(row) for row in zip(*array)][::-1]
````

## Problem 3

Write a function that takes a string as input and returns a dictionary containing the number of occurrences of each character in the string. Use method `count` of strings. Implement this function in one line using $\lambda$-function.

Example for the string 'hello, world!': {'h': 1, 'e': 1, 'l': 3, 'o': 2, ',': 1, ' ': 1, 'w': 1, 'r': 1, 'd': 1, '!': 1}.

````python

func = lambda string: {char: string.count(char) for char in string}
print(func('hello, world!'))
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
from datetime import datetime


class Book:
    def __init__(self, title, author, year, isbn, copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies

    def info(self):
        return f'"{self.title}" by {self.author} ({self.year}), ISBN: {self.isbn}, Copies: {self.copies}'

    def change_copies(self, amount):
        if self.copies + amount < 0:
            raise ValueError("Not enough copies available.")
        self.copies += amount


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []

    def add_book(self, book):
        self.borrowed_books.append(book)

    def remove_book(self, book):
        self.borrowed_books.remove(book)

    def info(self):
        books = [b.title for b in self.borrowed_books] or ["none"]
        return f"{self.name} (Card: {self.card_number}), Borrowed: {', '.join(books)}"


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.history = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, isbn):
        self.books = [b for b in self.books if b.isbn != isbn]

    def add_user(self, user):
        self.users.append(user)

    def remove_user(self, card_number):
        self.users = [u for u in self.users if u.card_number != card_number]

    def find_books(self, **criteria):
        result = self.books
        for key, value in criteria.items():
            result = [b for b in result if str(getattr(b, key, "")).lower() == str(value).lower()]
        return result

    def find_users(self, **criteria):
        result = self.users
        for key, value in criteria.items():
            result = [u for u in result if str(getattr(u, key, "")).lower() == str(value).lower()]
        return result

    def issue_book(self, card_number, isbn):
        user = next((u for u in self.users if u.card_number == card_number), None)
        book = next((b for b in self.books if b.isbn == isbn), None)
        if not user or not book:
            raise ValueError("User or book not found.")
        book.change_copies(-1)
        user.add_book(book)
        self.history.append({"action": "issued", "user": user.name, "book": book.title, "date": datetime.now()})
        print(f'Book "{book.title}" issued to {user.name}.')

    def return_book(self, card_number, isbn):
        user = next((u for u in self.users if u.card_number == card_number), None)
        book = next((b for b in self.books if b.isbn == isbn), None)
        if not user or not book:
            raise ValueError("User or book not found.")
        book.change_copies(1)
        user.remove_book(book)
        self.history.append({"action": "returned", "user": user.name, "book": book.title, "date": datetime.now()})
        print(f'Book "{book.title}" returned by {user.name}.')

    def show_history(self):
        print("\n--- Transaction History ---")
        for record in self.history:
            print(f"[{record['date'].strftime('%Y-%m-%d %H:%M:%S')}] "
                  f"{record['user']} {record['action']} \"{record['book']}\"")


# --- Demo ---
library = Library()

# Books
b1 = Book("1984", "George Orwell", 1949, "978-0451524935", 3)
b2 = Book("Dune", "Frank Herbert", 1965, "978-0441013593", 2)
b3 = Book("The Hobbit", "J.R.R. Tolkien", 1937, "978-0547928227", 5)
for b in [b1, b2, b3]:
    library.add_book(b)

# Users
u1 = User("Alice", "LIB-001")
u2 = User("Bob", "LIB-002")
for u in [u1, u2]:
    library.add_user(u)

# Issue & return
library.issue_book("LIB-001", "978-0451524935")
library.issue_book("LIB-002", "978-0441013593")
library.issue_book("LIB-001", "978-0547928227")
library.return_book("LIB-001", "978-0451524935")

# Search
print("\n--- Search by author 'Frank Herbert' ---")
for book in library.find_books(author="Frank Herbert"):
    print(book.info())

# Users info
print("\n--- Users ---")
for user in library.users:
    print(user.info())

# History
library.show_history()
````

Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python

A = 0.0
B = 0.0
for i in range(1,10001):
  A += 1 / (i**2)


for i in range(10000,0,-1):
  B += 1 / (i**2)

print(A)
print(B)

if A < B:
  print("A is lesser than B")
elif A > B:
  print("B is lesser than A")
else:
  print("Numbers are equal")
````
